```python
df = df.loc[(df['year']<=2023)& (df['year']>=1995)]
df.shape
```

```python
# EDA
df['brand'].unique()
```

```python
def segment_car(brand):
  if brand in ['audi','bmw','ferrari','jaguar','lamborghini','land-rover','volvo','mercedes-benz']:
    return 'top model'
  elif brand in ['chevrolet','fiat','honda','hyundai','nissan','ford','kia','toyota','porsche']:
    return 'medium model'
  else:
    return 'low model'
df['brand'] = df['brand'].apply(segment_car)
```

```python
df['voivodeship'].unique()
```

```python
def segment_group(voivodeship):
  if voivodeship in ['Łódzkie', 'Mazowieckie', 'Kujawsko-pomorskie', 'Podkarpackie','Wielkopolskie', 'Pomorskie', 'Małopolskie', 'Podlaskie']:
    return 'east poland'
  elif voivodeship in ['Śląskie', 'Dolnośląskie', 'Lubelskie', 'Świętokrzyskie','Zachodniopomorskie', 'Lubuskie', 'Opolskie']:
    return 'west poland'
  elif voivodeship in ['Warmińsko-mazurskie', 'Praga', 'Moravian-Silesian Region','Brandenburg', 'Rheinland-Pfalz', 'Anglia', 'Berlin', 'Hessen']:
    return 'north poland'
  else:
    return 'south poland'

df['voivodeship'] = df['voivodeship'].apply(segment_group)
```

```python
# split the data
X =df.drop('price_in_pln',axis= 1)
Y = df['price_in_pln']
```

```python
print(X)
```

```python
print(Y)
```

```python
# split train test data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state = 0)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

dtree = DecisionTreeRegressor()
dtree.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
param_grid = {
    'max_depth':[2,4,6,8],
    'min_samples_split':[2,4,6,8],
    'min_samples_leaf':[1,2,3,4],
    'max_features':['auto','sqrt','log2'],
    'random_state':[0,42]
}

grid_search = GridSearchCV(dtree,param_grid,cv=5,scoring ='neg_mean_squared_error')

grid_search.fit(X_train,Y_train)
print(grid_search.best_params_)
```

```python
dtree =DecisionTreeRegressor(max_depth=8,max_features='auto',min_samples_leaf = 1,min_samples_split=8,random_state=42)

dtree.fit(X_train,Y_train)
```

```python
import math
from sklearn.metrics import mean_absolute_percentage_error
```

```python
from sklearn import metrics
```

```python
Y_pred = dtree.predict(X_test)
mse = metrics.mean_squared_error(Y_test,Y_pred)
r2 = metrics.r2_score(Y_test,Y_pred)
rmse = math.sqrt(mse)
```

```python
print('MSE is:',mse)
print('R2 score is:',r2)
print('RSME is :',rmse)
```
